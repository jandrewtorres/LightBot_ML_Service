## Create pandas dataframe from user ActionHistory

```python
df = pd.DataFrame.from_dict(actionHistory)
df
```

```python
intent_col = df['intent'].values
color_col = df['intentColor'].values

def intentToCategory(intent):
    if(intent == 'light_on'):
        return 1
    else:
        return 0

def colorToCategory(color):
    if(color == 'none'):
        return 0

new_intent_col = [intentToCategory(intent) for intent in intent_col]
new_color_col = [colorToCategory(color) for color in color_col]

df = df.drop(['intent', 'intentColor', '_id'], 1)

df['intent'] = pd.DataFrame(new_intent_col)
df['intentColor'] = pd.DataFrame(new_color_col)

df
```

```python
x_col_names = ['intent', 'intentColor', 'dayOfMonth', 'dayOfWeek']
X = df[x_col_names]
X
```

```python
y_col_names = ['hour', 'minutes']
y = df[y_col_names] 
y
```

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0)
```

```python
ytrain
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(max_features='log2', n_estimators=150)
model.fit(Xtrain, ytrain)
model.score(Xtest, ytest)
```

```python
ytestpred = model.predict(Xtest)
ytrainpred = model.predict(Xtrain)
```

```python
print('Score: ', model.score(Xtest, ytest))
ytrainpred = pd.DataFrame(ytrainpred)

newytrainpred = pd.DataFrame(columns=['hour', 'minutes'])
def roundRow(row):
    hour = round(row[0])
    minute = round(row[1])
    return hour, minute

for index, row in ytrainpred.iterrows():
    hour, minute = roundRow(row)
    newytrainpred.at[index, 'hour'] = hour
    newytrainpred.at[index, 'minutes'] = minute
newytrainpred
```

```python
ytrain = ytrain.reset_index(drop=True)
pd.DataFrame(ytrain)
```

```python
import matplotlib.pyplot as plt

def toMinutes(row):
    return row['hour'] * 60 + row['minutes']

predicted = [toMinutes(row) for index, row in newytrainpred.iterrows()]
actual = [toMinutes(row) for index, row in ytrain.iterrows()]
x_axis = np.arange(0, 7, 1)
plt.figure()
plt.plot(x_axis, actual, label='actual')
plt.plot(x_axis, predicted, label='predicted')
plt.title('Actual action times vs. Predicted action times')
plt.xlabel('Arbitrary action index number')
plt.ylabel('Time of Day (minutes)')
plt.legend()
plt.show()
```

## Conclusions

### What worked:
- It predicted the time well
- It is very simple

### What didn't work:
- If the intent is used at different times throughout the day, you will end up with bad values. For example, if the test user turned the light on and off in the morning as well, our regression model would give an average value and would be no good for our purposes.

### How it could work:
- If we made sure that intents were unique to actions: for example 'going to bed' only happens once per day (usually), then this would work well.

### The problem:
- It is unlikely that users of LightBot will only perform actions only once a day. For example: you go to class more than once in a day. How would LightBot differentiate between actions? We would have to do a lot of filtering. This would require a lot of extra overhead that we do not have time to implement.
- This is overkill. If we did have unique intents, we could just take the average times and put that into the user schedule instead of implementing a machine learning service. This is essentially a really fancy model for getting average on/off times.
